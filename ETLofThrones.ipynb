```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
battles_file = "data/battles.csv"
battles_df = pd.read_csv(battles_file)
battles_df.head()
```

```python
deaths_file = "data/character-deaths.csv"
deaths_df = pd.read_csv(deaths_file)
deaths_df.head(100)
```

```python
predictions_file = "data/character-predictions.csv"
predictions_df = pd.read_csv(predictions_file)
predictions_df.head()
```

```python
# Specify which columns from deaths_df we want for characters table
deaths_cols = ["Name", "Allegiances", "Gender", "GoT", "CoK", "SoS", "FfC", "DwD"]
deaths_transformed = deaths_df[deaths_cols].copy()

# Rename the columns 
deaths_transformed = deaths_transformed.rename(columns={"Name": "CharacterName",
                                                        "Allegiances": "House",
                                                        "Gender": "Gender",
                                                        "GoT": "Book1",
                                                        "CoK": "Book2",
                                                        "SoS": "Book3",
                                                        "FfC": "Book4",
                                                        "DwD": "Book5"
                                                        })

deaths_transformed.set_index("CharacterName", inplace=True)
deaths_transformed.head(100)
```

```python
# Specify which columns from deaths_df we want
battles_cols = ["name", "battle_number", "year", "attacker_1", "defender_1", "attacker_outcome"]
battles_transformed = battles_df[battles_cols].copy()

# Rename the columns 
battles_transformed = battles_transformed.rename(columns={"name": "BattleName",
                                                          "battle_number": "Battle_ID",
                                                          "year": "Battle_Year",
                                                          "attacker_1": "Attacker",
                                                          "defender_1": "Defender",
                                                          "attacker_outcome": "Attacker_Outcome"
                                                        })
battles_transformed.set_index("BattleName", inplace=True)
battles_transformed.head(100)
```

```python
# Specify which columns from deaths_df we want
predictions_cols = ["name", "popularity"]
predictions_transformed = predictions_df[predictions_cols].copy()

# Rename the columns 
predictions_transformed = predictions_transformed.rename(columns={"name": "CharacterName",
                                                                  "popularity": "Popularity"
                                                                 })
predictions_transformed.set_index("CharacterName", inplace=True)
predictions_transformed.head(100)
```

```python
# Specify which columns from deaths_df we want for characters table
deaths2_cols = ["Name", "Death Year", "Book of Death"]
deaths2_transformed = deaths_df[deaths2_cols].copy()

# Rename the columns 
deaths2_transformed = deaths2_transformed.rename(columns={"Name": "CharacterName",
                                                        "Death Year": "DeathYear",
                                                        "Book of Death": "Book_of_Death"
                                                        })
deaths2_transformed = deaths2_transformed.dropna(subset=['Book_of_Death'])

deaths2_transformed.set_index("CharacterName", inplace=True)
deaths2_transformed.head(100)
```
