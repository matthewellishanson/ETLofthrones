```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
battles_file = "data/battles.csv"
battles_df = pd.read_csv(battles_file)
battles_df.head()
```

```python
deaths_file = "data/character-deaths.csv"
deaths_df = pd.read_csv(deaths_file)
deaths_df.head()
```

```python
predictions_file = "data/character-predictions.csv"
predictions_df = pd.read_csv(predictions_file)
predictions_df.head()
```

```python
# Specify which columns from deaths_df we want
deaths_cols = ["Name", "Allegiances", "Death Year", "Book of Death", "Gender", "GoT", "CoK", "SoS", "FfC", "DwD"]
deaths_transformed = deaths_df[deaths_cols].copy()

# Rename the columns 
deaths_transformed = deaths_transformed.rename(columns={"Name": "CharacterName",
                                                        "Allegiances": "House",
                                                        "Death Year": "DeathYear",
                                                        "Book of Death": "Book_of_Death",
                                                        "Gender": "Gender",
                                                        "GoT": "Book1",
                                                        "CoK": "Book2",
                                                        "SoS": "Book3",
                                                        "FfC": "Book4",
                                                        "DwD": "Book5"
                                                        })

deaths_transformed.head(100)
```

```python
# Specify which columns from deaths_df we want
battles_cols = ["name", "battle_number", "year", "attacker_1", "defender_1", "attacker_outcome"]
battles_transformed = battles_df[battles_cols].copy()

# Rename the columns 
battles_transformed = battles_transformed.rename(columns={"name": "BattleName",
                                                          "battle_number": "Battle_ID",
                                                          "year": "Battle_Year",
                                                          "attacker_1": "Attacker",
                                                          "defender_1": "Defender",
                                                          "attacker_outcome": "Attacker_Outcome"
                                                        })
battles_transformed.head(100)
```
